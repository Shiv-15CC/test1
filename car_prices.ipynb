```python
#y_prec=array_data@hypothesis
#print(y_prec)
y=y.values
y_standardized=(y-y.mean())/y.std()
y_actual = y_standardized.reshape(-1, 1)
y_actual
```

```python
def gradient_descent(array_data,hypothesis,y_actual,learning_rate,iterations):
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        y_prec=array_data.dot(hypothesis)
        gradients = (1 / 23) * np.dot(array_data.T, (y_prec - y_actual))
        hypothesis=hypothesis-(learning_rate*gradients)
        cost_history[i]=(1/(2*23)) * np.sum((y_prec - y_actual)**2)
    return hypothesis,cost_history
    
```

```python
iterations=3500
learning_rate=0.003
hypothesis,cost_history=gradient_descent(array_data,hypothesis,y_actual,learning_rate,iterations)
hypothesis
cost_history
```

```python
import matplotlib.pyplot as plt

plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Over Iterations")
plt.show()
```
